```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import math
```

```python
response = requests.get('https://www.transcustoms.com/HS_tree.htm')
```

```python
response.encoding = 'utf-8'
```

```python
response.text
```

```python
if response:
    content = response.text
    soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

```python
soup.find_all('a', target="_blank")
```

```python
links = []
```

```python
for link in soup.find_all('a', target="_blank"):
    links.append(link.get('href'))
```

```python
len(links)
```

```python
links[991]
```

```python
with open("heading_links.txt", "w", encoding="utf-8") as file:
    for item in links:
        file.write(str(item) + "\n")  # Chuyển sang chuỗi và ghi vào file
```
