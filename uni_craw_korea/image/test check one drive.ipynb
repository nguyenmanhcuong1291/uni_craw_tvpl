```python


def get_all_files_in_folder(access_token, folder_path):
    headers = {
        "Authorization": f"Bearer {access_token}"
    }

    # URL để truy cập thư mục
    url = f"https://graph.microsoft.com/v1.0/me/drive/root:/{folder_path}:/children"
    all_files = []

    while url:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Kiểm tra lỗi từ response

        # Lấy dữ liệu JSON từ phản hồi
        data = response.json()
        print(data)
        items = data.get('value', [])
        
        for item in items:
            # Lưu thông tin file và URL
            file_info = {
                "Tên file": item['name'],
                "Loại": item['file']['mimeType'] if 'file' in item else 'Folder',
                "Link tải": item.get('@microsoft.graph.downloadUrl', 'Không có link tải')
            }
            all_files.append(file_info)
        
        # Kiểm tra phân trang
        url = data.get('@odata.nextLink', None)

    # Chuyển thành DataFrame
    df_files = pd.DataFrame(all_files)
    return df_files

# Đường dẫn thư mục trên OneDrive và access token
folder_path = r"Attachments/Code/uni_craw_tvpl/uni_craw_korea/image"  # Đường dẫn thư mục trên OneDrive

# Lấy danh sách toàn bộ file trong thư mục
df_all_files = get_all_files_in_folder(access_token, folder_path)

# Lưu DataFrame ra file Excel
output_path = r"D:\Onedrive\OneDrive - UNI CONSULTING\Attachments\Code\uni_craw_tvpl\uni_craw_korea\image\all_files_links.xlsx"
df_all_files.to_excel(output_path, index=False)

print("Hoàn thành lấy danh sách toàn bộ file và link tải trong thư mục!")
```
