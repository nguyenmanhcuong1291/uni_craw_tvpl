```python
import requests
from bs4 import BeautifulSoup
import re
import os
import pandas as pd
from lxml import etree
from datetime import datetime
import openpyxl
from openpyxl import Workbook
from openpyxl import load_workbook
current_directory = os.getcwd()
```

```python
url_template = "https://thuvienphapluat.vn/page/tim-van-ban.aspx?keyword=m%C3%A3+s%E1%BB%91&area=0&match=True&type=21&status=0&signer=0&sort=1&lan=1&scan=0&org=0&fields=4&page={}"


def get_links_from_page(page_number):
    url = url_template.format(page_number)
    response = requests.get(url)
    

    if response.status_code != 200:
        print(f"Lỗi khi tải trang {page_number}")
        return []


    soup = BeautifulSoup(response.text, 'html.parser')


    divs = soup.find_all('div', class_=re.compile(r'content-[01]'))


    links = [div.find('a')['href'] for div in divs if div.find('a')]
    
    return links
```

```python
all_links = []
page = 1  
while True:
    try:
        print(f"Đang tải trang {page}")
        links = get_links_from_page(page)  
        count = len(links)
        print(f"Thu thập được {count} link")

        if count == 0:  # không còn link dừng vòng lặp
            break 

        all_links.extend(links)
        print(f"Tổng số link thu thập được: {len(all_links)}")

        page += 1  
    except Exception as e:
        print(f"Lỗi xảy ra: {e}")
        continue  
```

```python
with open('links_18022025.txt', 'w', encoding='utf-8') as f:
    for link in all_links:
        f.write(link + '\n')
```

```python
with open("links_old.txt", "r", encoding="utf-8") as links_old:
    old_links = set(links_old.read().splitlines())

with open("links_18022025.txt", "r", encoding="utf-8") as links_new:
    new_links = set(links_new.read().splitlines())

links_for_craw = new_links - old_links
print(len(links_for_craw))

with open("links_for_craw.txt", "w", encoding="utf-8") as f_out:
    for link in links_for_craw:
        f_out.write(link + "\n")
```
