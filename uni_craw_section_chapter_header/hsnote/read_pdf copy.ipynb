```python
import pytesseract
from pdf2image import convert_from_path
from PIL import Image, ImageDraw, ImageFont
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
import os

# üîπ C·∫•u h√¨nh ƒë∆∞·ªùng d·∫´n
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
poppler_path = r"C:\poppler-24.08.0\Library\bin"

# üîπ ƒê·ªçc file PDF scan v√† chuy·ªÉn th√†nh ·∫£nh
pdf_path = "MAWB (828-56028206).pdf"
images = convert_from_path(pdf_path, poppler_path=poppler_path)

# üîπ L∆∞u ·∫£nh v√† OCR vƒÉn b·∫£n v√†o TXT
text_file_path = "output_text.txt"
image_paths = []

with open(text_file_path, "w", encoding="utf-8") as f:
    for i, img in enumerate(images):
        image_path = f"page_{i+1}.png"
        img.save(image_path)  # L∆∞u ·∫£nh g·ªëc ƒë·ªÉ x·ª≠ l√Ω sau
        image_paths.append(image_path)

        text = pytesseract.image_to_string(img, lang="eng")  # Nh·∫≠n di·ªán vƒÉn b·∫£n
        f.write(f"--- PAGE {i+1} ---\n{text}\n\n")

print(f"‚úÖ VƒÉn b·∫£n ƒë√£ ƒë∆∞·ª£c l∆∞u v√†o {text_file_path}. H√£y ch·ªânh s·ª≠a n√≥ r·ªìi ch·∫°y b∆∞·ªõc ti·∫øp theo!")


```

```python
# ------------------------------------------
# üîπ ƒê·ªçc l·∫°i file TXT sau khi ch·ªânh s·ª≠a
with open(text_file_path, "r", encoding="utf-8") as f:
    edited_text = f.read().split("--- PAGE ")

# üîπ Ghi ƒë√® vƒÉn b·∫£n m·ªõi l√™n ·∫£nh g·ªëc v√† l∆∞u th√†nh PDF
output_pdf_path = "edited_document.pdf"
c = canvas.Canvas(output_pdf_path, pagesize=letter)

for i, image_path in enumerate(image_paths):
    img = Image.open(image_path)
    draw = ImageDraw.Draw(img)
    font = ImageFont.load_default()  # D√πng font m·∫∑c ƒë·ªãnh

    page_text = edited_text[i+1].strip() if i+1 < len(edited_text) else ""
    
    y_position = 50  # V·ªã tr√≠ vi·∫øt vƒÉn b·∫£n m·ªõi
    for line in page_text.split("\n"):
        draw.text((50, y_position), line, font=font, fill=(0, 0, 0))
        y_position += 20  # Xu·ªëng d√≤ng
    new_image_path = f"edited_page_{i+1}.png"
    img.save(new_image_path)
    new_image_path = f"edited_page_{i+1}.png"
    img.save(new_image_path)  # L∆∞u ·∫£nh ƒë√£ ch·ªânh s·ª≠a

    # üîπ Th√™m ·∫£nh ƒë√£ ch·ªânh s·ª≠a v√†o PDF
    c.drawImage(new_image_path, 0, 0, width=612, height=792)  # Fit k√≠ch th∆∞·ªõc trang PDF
    c.showPage()

c.save()

print(f"‚úÖ File PDF m·ªõi ƒë√£ ƒë∆∞·ª£c t·∫°o: {output_pdf_path}")
```
