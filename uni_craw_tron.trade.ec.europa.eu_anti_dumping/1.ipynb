```python
import requests

url = "https://tron.trade.ec.europa.eu/investigations/api/eucase/search"

headers = {
    "Accept": "application/json; charset=UTF-8",
    "Accept-Language": "en-US,en;q=0.9,vi;q=0.8,vi-VN;q=0.7",
    "Access-Control-Allow-Origin": "*",
    "Cache-Control": "no-cache",
    "Connection": "keep-alive",
    "Content-Type": "application/json; charset=UTF-8",
    "DNT": "1",
    "Origin": "https://tron.trade.ec.europa.eu",
    "Pragma": "no-cache",
    "Referer": "https://tron.trade.ec.europa.eu/investigations/search",
    "Sec-Fetch-Dest": "empty",
    "Sec-Fetch-Mode": "cors",
    "Sec-Fetch-Site": "same-origin",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36",
    "sec-ch-ua": '"Not A(Brand";v="8", "Chromium";v="132", "Google Chrome";v="132"',
    "sec-ch-ua-mobile": "?0",
    "sec-ch-ua-platform": '"Windows"',
}

data = {
    "cncode": "",
    "casenumber": "",
    "productname": "",
    "country": "All countries",
    "measure": "Measures in force",
    "measuretype": "All types of measure",
    "status": "Any status",
    "proceeding": "Anti-dumping",
}

response = requests.post(url, headers=headers, json=data)

print(response.status_code)
data = response.json()  # In kết quả JSON trả về
```

```python
data = response.json() 
```

```python
import pandas as pd 
```

```python
data_df = pd.DataFrame(data)
```

```python
data_df['Countries investigated'] = data_df['caseCountries'].apply(lambda x: [i['countryName'] for i in x] if len(x) > 0 else None)
```

```python
data_df['Related case numbers'] = data_df['relatedCases'].apply(
    lambda x: ['Ongoing: ' + i['caseNumber'] if i['ongoing'] else i['caseNumber'] for i in x] if len(x) > 0 else None
)
```

```python
data_df
```

```python
final_df = data_df[['shortName', 'Countries investigated', 'caseCategory' ,'measureStatus', 'ongoingInvestigation', 'Related case numbers']] 
```

```python
final_df['Countries investigated'] = final_df['Countries investigated'].apply(lambda x :"\n".join(x) if x is not None else None)
```

```python
final_df['Related case numbers'] = final_df['Related case numbers'].apply(lambda x :"\n".join(x) if x is not None else None)
```

```python
final_df.columns = ['Product', 'Countries investigated', 'Proceeding', 'Measure', 'Status', 'Related case numbers']
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Alignment

# Xuất DataFrame ra file Excel
final_df.to_excel("EU_Trade.xlsx", index=False, engine="openpyxl")

# Mở file Excel để chỉnh sửa định dạng
wb = load_workbook("EU_Trade.xlsx")
ws = wb.active

# Áp dụng wrap text cho tất cả các ô
for row in ws.iter_rows():
    for cell in row:
        cell.alignment = Alignment(wrap_text=True)

# Lưu lại file Excel với định dạng xuống dòng
wb.save("EU_Trade.xlsx")
```
