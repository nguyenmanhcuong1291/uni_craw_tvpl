```python
from collections import Counter

# Đọc tất cả các dòng trong file
with open('eu_list.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Đếm số lần xuất hiện của từng dòng
line_counts = Counter(lines)

# Lọc và in ra các dòng lặp lại (số lần xuất hiện > 1)
for line, count in line_counts.items():
    if count > 1:
        print(f"'{line.strip()}' xuất hiện {count} lần")
```
